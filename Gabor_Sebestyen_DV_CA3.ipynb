## Q2.

### My first visualization is aiming the 17-35-year group, or with other words the Young Adults. I used line plot because I was tasked to visualize the value change of a group of variables within a period of time and line plot is well suitable for this purpose (Ed.gov., 2019). 
### To reach the attention of this age group I aimed to be explicit, however also have things handy. I also tried to be straight forward and not to over complicate the visualization. I used the appropriate title and labeling and legend presentation in an aim to aid the simplicity and the proper understanding. I also used bright palette to draw attention to the graph. I also decided to use the gridding in an aim to aid the better understanding of the values.


### I generated a line plot to visualize the change of values in the High variable for all Currencies over time (Data Viz, 2020), (Stack Overflow, n.d.), (Walker, 2019),  .

```python
sns.set(rc={'figure.figsize':(18,14)})
ax = sns.lineplot(data=currency_high, x="Date", y="High", hue="Currency", palette="bright", linewidth = 4)
ax.legend(bbox_to_anchor=(1, 0.9),borderaxespad=0, fancybox=True, framealpha=4, shadow=True, borderpad=3)
ax.legend(fontsize=16)

plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2 , axis='y')

plt.title("Change of high value of Cryptocurrencies from 2013 to 2019")
plt.xlabel("Date")
plt.ylabel("High value")
plt.show()
```

### My visualization for the age group of 60+, so called the Seniors was generated with the similar approach as the previous one for the other age group. However, in this case to try to aid a Senior- friendly platform, I used relplot. With this I was able to present the change of the values of the High variable separately as per relevant Currency, yet still was able to present this on the same visualization. I used relplot because I was able to present the change of values in a period of time for multiple variables on the same visualization. 
### This way the presentation was clear and the changes were easier observed. I also manually selected the line colors to be more senior friendly and without the blue color. Appropriate titles and labels were also applied. 


### I used relplot to generate my visualization. I used cloumn wrap to easy on the observation. I also rotated my x label for the easier read and better presentation (seaborn.pydata.org, n.d.), (Stack Overflow, n.d.). 

```python
with sns.plotting_context(rc={"legend.fontsize":16, "font.size":16, "label.size":16}):
    sns.set(rc={'figure.figsize':(12,10)})
    ax=sns.relplot(
        data=currency_high, x="Date", y="High",
        col="Currency", col_wrap=3, hue="Currency",
        markers=True, kind="line", palette=["r","g","y","#800080","#00FF00","#FF00FF","#808000","#bfff00","#ff0080","#bf00ff","#ac5353","#ffb3b3"], linewidth = 4)
    ax.set_xticklabels(rotation=30)
    ax.fig.suptitle("Change of high value of Cryptocurrencies from 2013 to 2019")
    ax.fig.subplots_adjust(top=0.95)
```

### I removed the commas from the format of the Volume variable and then converted the values of the Volume variable to float data type (pbpython.com., n.d.) , so I could apply the appropriate visualization. I also made a number of observations to make sure the necessary changes were done.

### I generated a catplot to see which values of the Currency variable will be required for my next task.

### I made an observation of the unique values of the Currency variable, this was literally for the purpose of easier reading of the name of the values. I then removed all irrelevant Currency values so my dataset only shall have the necessary ones.

```python
currency_volume.Currency.unique()
```

```python
currency_volume = currency_volume[currency_volume.Currency != 'binance-coin']
currency_volume = currency_volume[currency_volume.Currency != 'eos']
currency_volume = currency_volume[currency_volume.Currency != 'tezos']
currency_volume = currency_volume[currency_volume.Currency != 'xrp']
currency_volume = currency_volume[currency_volume.Currency != 'litecoin']
currency_volume = currency_volume[currency_volume.Currency != 'bitcoin-cash']
currency_volume = currency_volume[currency_volume.Currency != 'bitcoin-sv']
currency_volume = currency_volume[currency_volume.Currency != 'stellar']
currency_volume = currency_volume[currency_volume.Currency != 'cardano']
```

### My first visualization is again aiming the Young Adults. Due to the same reasons as at the previous task my approach was based on line plot. Beside the benefit of displaying the range of multiple values we are also able to observe their minimum and maximum (Studypug.com., 2019). For the 3 data values we can receive perfect observation conditions as this plot is suitable for up to 50 data values.
### The tools I used to get the appropriate attention of the relevant group is similar to what I used at my previous line plot. The appropriate colors, grid and necessary labelling. This time I also disabled the scientific displaying for large values (Stack Overflow., n.d.).. 

```python
sns.set(rc={'figure.figsize':(18,14)})
ax = sns.lineplot(data=currency_volume, x="Date", y="Volume", hue="Currency", palette="bright", linewidth = 2)
ax.get_yaxis().get_major_formatter().set_scientific(False)
# https://stackoverflow.com/questions/14711655/how-to-prevent-numbers-being-changed-to-exponential-form-in-python-matplotlib-fi

plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2 , axis='y')

ax.legend(bbox_to_anchor=(1, 0.9),borderaxespad=0)
ax.legend(fancybox=True, framealpha=4, shadow=True, borderpad=1)
ax.legend(fontsize=16)
plt.title("Change of volume of Cryptocurrencies from 2016 to 2019")
plt.xlabel("Date")
plt.ylabel("Volume of Currencies")
plt.show()
```

### Similarly, as with the previous plot for the Senior age group I again used relplot. I made an attempt to generate other different visualizations, however this was the most suitable one for the purpose. Other visualizations displayed unclear plots. The separation for the visualization of the values are giving a very clear picture of their changes and ranges.
### Here I also applied the appropriate colors without blue and larger font size for the legend (GitHub., n.d.). I again used larger lines for the plot so it will be easier to see. I also rotated the labels for the x axle for the easier read.


```python
with sns.plotting_context(rc={"legend.fontsize":16, "font.size":16, "label.size":16}):
    ax=sns.relplot(
        data=currency_volume, x="Date", y="Volume",
        col="Currency", col_wrap=3, hue="Currency",
        markers=True, kind="line", palette=["r","g","#ff8000"], linewidth = 2)
    ax.set_xticklabels(rotation=30)
    ax.fig.suptitle("Change of volume of Cryptocurrencies from 2016 to 2019")
    ax.fig.subplots_adjust(top=0.8)

    leg = ax._legend
    leg.set_bbox_to_anchor([1,0.7])
```
